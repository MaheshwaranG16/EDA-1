```python
house_of_cards = df[df['title'] == 'House of Cards']

if not house_of_cards.empty:
    show_id = house_of_cards['show_id'].values[0]
    director = house_of_cards['director'].values[0]
    print("Show ID:", show_id)
    print("Director:", director)
else:
    print("Show 'House of Cards' not found.")
```

```python
release_counts = df.groupby('release_year')['type'].count()
release_counts
```

```python
max_year = release_counts.idxmax()
max_releases = release_counts.max()
```

```python
plt.figure(figsize=(20, 6))
release_counts.plot(kind='bar', color='skyblue')
plt.title('Number of TV Shows and Movies Released Each Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Releases')
plt.xticks(rotation=45)
plt.axvline(x=max_year, color='red', linestyle='--', linewidth=2, label=f'Highest: {max_year} ({max_releases} releases)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
r_rating_tv_shows_count = df[(df['type'] == 'TV Show') & (df['rating'] == 'TV-MA') & (df['release_year'] > 2018)].shape[0]
print(f"The number of TV shows with 'R' rating after year 2018: {r_rating_tv_shows_count}")
```

```python
df['duration_minutes'] = df['duration'].str.extract('(\d+)', expand=False).astype(float)
# Multiply the duration by 60 for TV Shows
df.loc[df['type'] == 'TV Show', 'duration_minutes'] *= 20 * 60  # Assuming 1 season is approximately 20 episodes and each episode is 60 minutes

# Find the maximum duration
max_duration = df['duration_minutes'].max()

print(f"The maximum duration of a movie/show on Netflix is: {max_duration} minutes")
```

```python
tv_shows_df = df[df['type'] == 'TV Show']

# Count the occurrences of each country
country_counts = tv_shows_df['country'].value_counts()

# Get the country with the highest number of TV shows
country_with_highest_tv_shows = country_counts.idxmax()
highest_tv_show_count = country_counts.max()

print(f"The country with the highest number of TV shows is {country_with_highest_tv_shows} with {highest_tv_show_count} TV shows.")
```
